```python
import pandas as pd
personas = pd.read_csv("personas_2011.csv", sep=";")
conicet = pd.read_csv("ref_categoria_conicet.csv", sep=";")
# ctrl + shift + P : jupyter #para crear jupyter notebook(ipynb)
personas

# en iris.describe() 50% = la mediana
```

```python
pers = personas[["persona_id", "anio", "categoria_conicet_id"]]
pers.head(10)
```

```python
conicet.head(10)
```

```python
pd.merge(pers, conicet,how="right", on="categoria_conicet_id") #right prioriza la tabla de la derecha(conicet) o sea que cuando no aparezca un dato en conicet no va a verse

# """ 
# on: indica en base a quÃ© columna se van a unir los data frames. Esta columna tiene que estar presente en ambos data frames.

# left_on: nombre de la columna de los datos de la izquierda el cual se va a usar para unir los data frames."""
```

```python
pd.merge(pers, conicet,how="left", on = "categoria_conicet_id")
```

```python
pd.merge(pers, conicet,how="outer", on="categoria_conicet_id")
```

```python
pd.merge(pers, conicet,how="inner", on="categoria_conicet_id")
```
